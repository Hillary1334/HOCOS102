```python
class General(object):

        # __init__ is known as the constructor
        def __init__(self,a,b,c):
            self.a = a
            self.b = b
            self.c = c
            
        def display1(self):
            print(self.a)
            print(self.b)
            print(self.c)
            print(self.d)
            print(self.e)
            print(self.f)
            print(" ")

        def display3(self):
            print(self.a)
            print(self.b)
            print(self.c)
            print(self.l)
            print(" ")

        def display2(self):
            print(self.a)
            print(self.b)
            print(self.c)
            print(self.g)
            print(self.h)
            print(self.i)
            print(self.j)
            print(self.k)
            print(" ")
            
# child class
class Retail(General):
    def __init__(self,a,b,c,d,e,f):    
        self.d = d
        self.e = e
        self.f = f

        # invoking the __init__ of the parent class
        General.__init__(self,a,b,c)

class Global(General):
    def __init__(self,a,b,c,g,h,i,j,k):    
        self.g = g
        self.h = h
        self.i = i
        self.j = j
        self.k = k

        # invoking the __init__ of the parent class
        General.__init__(self,a,b,c)

class Commercial(General):
    def __init__(self,a,b,c,l):    
        self.l = l

        # invoking the __init__ of the parent class
        General.__init__(self,a,b,c)

                
# creation of an object variable or an instance
aa = Retail("Lines of credit","Investment management and accounts","Insurance",
           "Retirement and education accounts","loans and mortgages","Checking and saving")
bb = Global("Lines of credit","Investment management and accounts","Insurance",
           "Multi-currency management services and products","Foreign currency accounts","Foreign currency credit cards",
           "Transborder advisory services","Liquidity management")
cc = Commercial("Lines of credit","Investment management and accounts","Insurance","Advisory services")

# calling a function of the class Person using its instance
aa.display1()
bb.display2()
cc.display3()
```
