```python
import random
import tkinter as tk
from tkinter import messagebox


class varchecker():
    
     def __init__(self, name):
         self.name = name

     def checker(self):
         name = name_entry.get()
         task = random.choice(Tasks)
         present.append(name)
         
         if name in Employees:
             messagebox.showinfo("ACCESS GRANTED", f"WELCOME, {name}: Your task today is: " + str(task))
         else:
             messagebox.showinfo("INVALID USER","USER DATA NOT FOUND, PLEASE CONFIRM DETAILS AND TRY AGAIN")

     def attendance(self):
         name = name_entry.get()
         if name in Employees:
             window2 = tk.Toplevel(root)
             listbox = tk.Listbox(window2,height = 20,width =150)
             window2.geometry("300x250")
             label = tk.Label(window2, text = "ATTENDANCE")
             for i in present:
                 listbox.insert(1,i)
             label.pack()
             listbox.pack()
         else:
             messagebox.showinfo("INVALID USER","USER DATA NOT FOUND, PLEASE CONFIRM DETAILS AND TRY AGAIN")
             
         
Employees = ["Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo",
"Abraham Jones" ,"Nicole Anide", "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"]

Tasks = ["Loading", "Transporting", "Reveiwing Orders", "Customer Service", "Delivering Items"]

present = []

root = tk.Tk()
root.title("login form")
root.geometry("500x200")

name_label = tk.Label(root, text = "PLEASE ENTER YOUR NAME")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

name = name_entry.get()
obj = varchecker(name)

submit_button = tk.Button(root, text="submit", command = obj.checker)
submit_button.pack()

submit_button = tk.Button(root, text="view attendance", command = obj.attendance)
submit_button.pack()

root.mainloop()



             
        
        

```
